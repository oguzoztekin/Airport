```python
import json
import pandas as pd
import os

# Load data from a JSON file
with open('SAW.json', 'r', encoding='utf-8') as file:
    airports = json.load(file)

# SAW airport details
column_a = "SAW (Turkey)"

column_a_list = []
column_b_list = []

l = []
for airport in airports:
    city = airport.get('city', '').title()
    country = airport.get('country', '').title()
    
    if country.lower() == 'turkey':
        continue  

    column_b = f"{city} ({country})"
    column_a_list.append(column_a)
    column_b_list.append(column_b)

# Create a DataFrame
df = pd.DataFrame({
    'Column A': column_a_list,
    'Column B': column_b_list
})

# Write the DataFrame to an Excel file
df.to_excel('output.xlsx', index=False)

```

```python
f = os.listdir()
key = list()
for ff in f:
    if '.json' in ff:
        key.append(ff)
```

```python
df = pd.DataFrame()
for f in key:
    # Load data from a JSON file
    try:
        with open(f'./{f}', 'r', encoding='utf-8') as file:
            airports = json.load(file)
        airports = pd.DataFrame(airports)
        airports['from'] = f.split('.')[0]
        df = pd.concat([df,airports])
    except: 
        continue
```
